```python
import pandas as pd

csv_data = pd.read_csv("shopping_data.csv")

# Membaca 5 Data Pertama
print("Menampilkan 5 data paling atas")
print(csv_data.head())

# Mengecek ada colom apa saja di data
print("Mengecek ada kolom apa saja di data")
print(csv_data.columns)

# Print kolom age nya saja
print("Menampilkan data bagian age nya saja")
print(csv_data['Age'])

# Print data menggunakan index colom
print("Menampilkan data menggunakan index kolom")
print(csv_data.iloc[5])

# Menampilkan suatu data dari baris dan kolom tertentu
print("Menampilkan suatu data dari baris dan kolom tertentu")
print(csv_data['Age'].iloc[1])

# Menampilkan data dalam range tertentu
print("Menampilkan data row age ke 5 sampai kurang dari 10 :")
print(csv_data['Age'].iloc[5:10])
print("Menampilkan data ke 5 sampai kurang dari 10 dalam satu baris:")
print(csv_data.iloc[5:10])

# Mengetahui Informasi statistik dari suatu data menggunakan fungsi describe
print("Informasi Statistik dari suatu data")
print(csv_data.describe(exclude=['O']))
```
